# Check known edges
check how many times edges have been checked for collisions to prove that edges are only checked once.

```python
import sys
from pathlib import Path
import numpy as np
from datetime import datetime

REPO_ROOT = Path.cwd().parent
sys.path.insert(0, str(REPO_ROOT))

REPO_ROOT

# Import custom modules
from planners.SBL import BidirectionalSBL
from modules import sbl_checkpoints as checkpoints
from modules import randomScene
```

```python
SCENE_LIMITS = np.array([[0,22],[0,22]])
CHECKPOINT_PATH = REPO_ROOT / "checkpoints" / "sbl_checkpoints.json"
GIF_PATH = REPO_ROOT / "gifs"
```

```python
for i in range(500):
    benchmark = randomScene.create_random_benchmark(SCENE_LIMITS)
    planner = BidirectionalSBL(benchmark.collisionChecker,{"count_edge_checks":True})
    T_start, T_goal, path = planner.plan_path(list(benchmark.startList[0]), list(benchmark.goalList[0]), checkpoint_path=str(CHECKPOINT_PATH))
    edge_check_counts = planner.collision_check_counter
    #print(f"{sum(edge_check_counts.values())} / {len(edge_check_counts)}")
    if np.any(np.array(list(edge_check_counts.values()))>1):
        raise RuntimeError("Edge was doublechecked for collision")
```
